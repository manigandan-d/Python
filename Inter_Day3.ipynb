```python
class Employee:
  empDept="Development"                #class attribute
  def __init__(self,fname,lname,sal):   #fname,lname,sal= local
    self.__fname=fname                   #instance attribute (private)
    self.__lname=lname
    self.__sal=sal

  def email(self):                     #email,fullname()  public
    return self.__fname+'.'+self.__lname+'@mail.com'

  def fullname(self):
    return self.__fname+' '+self.__lname



fname="Robin"
lname="Sharma"
sal=89000
emp1=Employee(fname,lname,sal)
#print(emp1.__fname)
print(emp1.email())
print(emp1.fullname())
print(Employee.empDept)
```

```python
class Employee:
  __empDept="Development"                #class attribute
  def __init__(self,fname,lname,sal,val):   #fname,lname,sal= local
    self.__fname=fname                   #instance attribute (private)
    self.__lname=lname
    self.__sal=sal
    self.sal=val

  def __email(self):                     #email - private
    return self.__fname+'.'+self.__lname+'@mail.com'

  def fullname(self):                    #fullname-public
    print(self.__email())
    return self.__fname+' '+self.__lname



fname="Robin"
lname="Sharma"
sal=89000
emp1=Employee(fname,lname,sal,40000)
#print(emp1.__fname)
#print(emp1.__email())
print(emp1.fullname())
print(emp1.sal)
#print(Employee.__empDept)
```

```python
#Closure:
Closure is a technique where your outer function holds the value even after going out of scope
```

```python
def expo(func):    #func=add  #4
  def inner(n1,n2):           #7  n1=10,n2=20
    return func(n1,n2)**2     #8   func-->add(10,20)  #11  30**2
  return inner    #5

def add(n1,n2):               #9    add(10,20)
  return n1+n2                #10   return 30

def prod(n1,n2):
  return n1*n2


no1=int(input("Enter No1:"))   #12
no2=int(input("Enter No2:"))   #2
colFun1=expo(add)              #3 expo(add)  #6=>ColFun1=inner
print("Power of sum:",colFun1(no1,no2))  #7  colFun1(10,20)   #12

colFun2=expo(prod)
print("Power of Pod:",colFun2(no1,no2))
```

```python
def expo(func):
  def inner(n1,n2):
    return func(n1,n2)**2
  return inner

@expo
def add(n1,n2):               #9    add(10,20)
  return n1+n2                #10   return 30

def prod(n1,n2):
  return n1*n2


no1=int(input("Enter No1:"))   #12
no2=int(input("Enter No2:"))   #2


print("Expo of Sum:",add(no1,no2))
print("Expo of Prod:",prod(no1,no2))
```
