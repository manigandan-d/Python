```python
class Point:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    def __add__(self, other):
        return Point(self.x + other.x, self.y + other.y)
    
    def __sub__(self, other):
        return Point(self.x - other.x, self.y - other.y)

    def __str__(self):
        return f"({self.x}, {self.y})"
    
p1 = Point(5, 2)
p2 = Point(4, 1)

print(p1)
print(p2)
print(p1 + p2)
print(p1 - p2)
```

```python
class Demo:
    def __init__(self, a):
        self.a = a

    def __add__(self, other):
        return self.a + other.a
    
    def __str__(self):
        return self.a
    
d1 = Demo(5)
d2 = Demo(3)
d3 = Demo(2)
d4 = Demo(1)

temp1 = d1 + d2
temp2 = Demo(temp1) + d3
d5 = Demo(temp2) + d4

print(d5)
```

```python
class ComplexNum:
    def __init__(self, real, img):
        self.__real = real
        self.__img = img
        
    def __str__(self):
        return f"{self.__real} + {self.__img}i"
    
    def __add__(self, other):
        real = self.__real + other.__real
        img = self.__img + other.__img
        
        # return f"{real} + {img}i"
        return ComplexNum(real, img)
        
comp1 = ComplexNum(3, 2)
comp2 = ComplexNum(5, 1)

print(comp1)
print(comp2)

comp3 = comp1 + comp2 
print(comp3)
```

```python
class Employee:
    def __init__(self, name, exp):
        self.__name = name
        self.__exp = exp 
        
    def __gt__(self, other):
        if self.__exp > other.__exp:
            return 1 
        
        else: 
            return 0
        
emp1 = Employee("mani", 25)
emp2 = Employee("anbu", 10)

if emp1 > emp2:
    print("Mani has higher experience")
    
else: 
    print("Anbu has higher experience")
```
