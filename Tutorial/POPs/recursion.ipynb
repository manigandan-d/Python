Recursion

Direct Recursion

```python
def fact(n):
    if n == 0:
        return 1

    else:
        return n * fact(n - 1)


n = 5
print(fact(n))
```

Tail Recursion

```python
def fun(n):
    if n == 0:
        return

    else:
        print(n, end=" ")

    return fun(n - 1)


n = 5
fun(n)
```

Non-tail Recursion

```python
def fact(n):
    if n == 0:
        return 1

    else:
        return n * fact(n - 1)


n = 5
print(fact(n))
```
