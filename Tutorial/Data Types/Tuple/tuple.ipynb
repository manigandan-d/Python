```python
userInfo = ("Jack", 30, "jack@123.com", 9182736405)

print(userInfo[0:2])
print(userInfo[1:])
print(userInfo[:3])
print(userInfo[0:5])

print(userInfo[::2])

print(userInfo[::-1])
```

operations on tuple

insertion

```python
date_time = (2024, 2, 14, 10, 30)

print(date_time)

temp = list(date_time)

temp.append(5)

date_time = tuple(temp)

print(date_time)
```

updation

```python
date_time = (2024, 2, 14, 10, 30, 5)

print(date_time)

temp = list(date_time)

temp[1] = 12
temp[2] = 25

date_time = tuple(temp)

print(date_time)
```

deletion

```python
date_time = (2024, 2, 14, 10, 30)

print(date_time)

temp = list(date_time)

temp.remove(2)
temp.remove(14)

date_time = tuple(temp)

print(date_time)
```

unpacking

```python
evenData = 2, 4, 6, 8, 10

e1, e2, e3, e4, e5 = evenData

print(e1)
print(e2)
print(e3)
print(e4)
print(e5)
```

usage of asterisk while unpacking

eg. 1

```python
evenData = 2, 4, 6, 8, 10

e1, e2, *e = evenData

print(e1)
print(e2)
print(e)
```

eg. 2

```python
evenData = 2, 4, 6, 8, 10

e1, *e, e5 = evenData
# *e, e2, e5 = evenData

print(e1)
print(e)
print(e5)
```
