keys()

```python
text = "hello"

char_count = {}

for char in text:
    char_count[char] = char_count.get(char, 0) + 1

char_count_keys = char_count.keys()

# print(char_count_keys)
print("Unique characters:", list(char_count_keys))

text = "hello world"

for char in text:
    char_count[char] = char_count.get(char, 0) + 1

# print(char_count_keys)
print("Unique characters:", list(char_count_keys))
```

copying a dictionary

shallow copy

```python
data = {"name": "John", "age": 30, "city": "chennai"}
print(data)

dupData = data
print(dupData)

print(id(data))
print(id(dupData))

dupData["age"] = 25

print(data)
print(dupData)
```

deep copy

copy()

```python
data = {"name": "John", "age": 30, "city": "chennai"}
print(data)

dupData = data.copy()
print(dupData)

print(id(data))
print(id(dupData))

dupData["age"] = 25

print(data)
print(dupData)
```

dict()

```python
data = {"name": "John", "age": 30, "city": "chennai"}
print(data)

dupData = dict(data)
print(dupData)

print(id(data))
print(id(dupData))

dupData["age"] = 25

print(data)
print(dupData)
```

dictionary comprehension

```python
data = {"Durai": 50, "Kalai": 40, "Keerai": 30, "Mani": 20, "Anbu": 10}

modData = {key: value + 10 for key, value in data.items()}
print(modData)
```

if statement in dictionary comprehension

```python
data = {"Durai": 50, "Kalai": 40, "Keerai": 30, "Mani": 20, "Anbu": 10}

modData = {key: value + 10 for key, value in data.items() if value > 30}
print(modData)
```

if else statement in dictionary comprehension

```python
data = {"Durai": 50, "Kalai": 40, "Keerai": 30, "Mani": 20, "Anbu": 10}

modData = {key: value + 10 if value > 30 else value + 5 for key, value in data.items()}
print(modData)
```
