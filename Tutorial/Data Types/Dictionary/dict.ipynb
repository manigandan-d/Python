```python
details = {"Name": "Mani", "Age": 22, "Gender": 'M'}

print(details["Name"])
print(details["Age"])
print(details["Gender"])
```

dictionary value as list, tuple, set & dictionary

```python
details = {"Name": "Mani", "Age": 22, "Gender": 'M', "Sib": ["Keerai", "Anbu"], "SibPosition": {"Keerai": "Computer Engineer", "Anbu": "Mechanical Engineer"}}

for key, value in details.items():
    print(key, end=":")
    print(value)
```

keys()

```python
details = {"Name": "Mani", "Age": 22, "Gender": 'M'}

print(details.keys())

for key in details.keys():
    print(key)
```

values()

```python
details = {"Name": "Mani", "Age": 22, "Gender": 'M'}

print(details.values())

for val in details.values():
    print(val)
```

items()

```python
details = {"Name": "Mani", "Age": 22, "Gender": 'M'}

print(details.items())

for key, val in details.items():
    print(key, end=":")
    print(val)
```

dictionary comprehension

```python
data = {"Durai": 50, "Kalai": 40, "Keerai": 30, "Mani": 20, "Anbu": 10}

modData = {key: value + 10 for key, value in data.items()}
print(modData)
```

if statement in dictionary comprehension

```python
data = {"Durai": 50, "Kalai": 40, "Keerai": 30, "Mani": 20, "Anbu": 10}

modData = {key: value + 10 for key, value in data.items() if value > 30}
print(modData)
```

if else statement in dictionary comprehension

```python
data = {"Durai": 50, "Kalai": 40, "Keerai": 30, "Mani": 20, "Anbu": 10}

modData = {key: value + 10 if value > 30 else value + 5 for key, value in data.items()}
print(modData)
```

```python
name = {"Durai", "Kalai", "Keerai", "Mani", "Anbu"}
age = {50, 40, 30, 20, 10}

zipped = dict((zip(name, age)))
print(zipped)
```
