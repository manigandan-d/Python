string formatting / string interpolation

f_string

```python
name = input("Enter your name: ")
age = int(input("Enter your age: "))
weight = float(input("Enter your weight: "))
height = float(input("Enter your height: "))

print(f"This is {name}. I'm {age} years old. My weight is {weight}kg. My height is {height}m.")

info = f"This is {name}. " \
       f"I'm {age} years old. " \
       f"My weight is {weight}kg. " \
       f"My height is {height}m."
       
print(info)

info = ( f"This is {name}. "
       f"I'm {age} years old. "
       f"My weight is {weight}kg. "
       f"My height is {height}m." )

print(info)

info = f"""This is {name}.
       I'm {age} years old.
       My weight is {weight}kg.
       My height is {height}m."""
       
print(info)
```

set precision

```python
from math import pi

golden_ratio = 1.61803
euler_constant = 2.71828
square_root_2 = 1.41421
square_root_3 = 1.73205

print(pi)
print("{:.2f}".format(pi))
print("%.2f" %(pi))
print("{:f}".format(3.14))
print("gr: {:f}, ec: {:f}, sqrt2: {:f} & sqrt3: {:f}".format(golden_ratio, euler_constant, square_root_2, square_root_3))
print("gr: {0:f}, ec: {1:f}, sqrt2: {2:f} & sqrt3: {3:f}".format(golden_ratio, euler_constant, square_root_2, square_root_3))
print("gr: {1:f}, ec: {3:f}, sqrt2: {0:f} & sqrt3: {2:f}".format(golden_ratio, euler_constant, square_root_2, square_root_3))
```
