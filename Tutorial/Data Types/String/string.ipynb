slicing

array slicing

```python
data = "python program"

print(data[1:4])

print(data[4:])
print(data[:4])

print(data[:])  # data[::]

print(data[::2])
print(data[1::2])
print(data[:7:2])

print(data[2:10:2])
print(data[8:1:-2])

print(data[::-1])
```

```python
data = "python program"

print(data[slice(1, 4)])

print(data[slice(4, 14)])
print(data[slice(4)])

print(data[slice(0, 14)])

print(data[slice(0, 14, 2)])

print(data[slice(0, 12, 2)])
print(data[slice(1, 14, 2)])
print(data[slice(0, 7, 2)])

print(data[slice(2, 10, 2)])
print(data[slice(8, 1, -2)])

# print(data[slice(,, -1)])
```

format()

```python
name = "Mani"
proLang1 = "C"
proLang2 = "C++"
proLang3 = "Python"
proLang4 = "Java"

print("{} knows {}, {}, {} & {}".format(name, proLang1, proLang2, proLang3, proLang4))
print("{0} knows {3}, {1}, {4} & {2}".format(name, proLang1, proLang2, proLang3, proLang4))
print("{n} knows {p1}, {p2}, {p3} & {p4}".format(n=name, p1=proLang1, p2=proLang2, p3=proLang3, p4=proLang4))
```
