if statement

```python
a = 10
b = 5

if a > b:
    print(f"{a} is greater than {b}")
```

short hand if 

```python
a = 10
b = 5

if a > b: print(f"{a} is greater than {b}")
```

else if

```python
a = 10
b = 5

if a > b: 
    print(f"{a} is greater than {b}")
    
else:
    print(f"{b} is greater than {a}")
```

short hand if else

```python
a = 10
b = 5

print(f"{a} is greater than {b}") if a > b else print(f"{b} is greater than {a}")
```

if elif else

```python
a = 5
b = 10
c = 15

if a > b and a > c:
    print("a is greater than b & c")

elif b > a and b > c:
    print("b is greater than a & c")

else:
    print("c is greater than a & b")
```

nested if

```python
a = 5
b = 10
c = 15

if a > b:
    if a > c:
        print("a is greater than b & c")

else:
    if b > c:
        print("b is greater than a & c")

    else:
        print("c is greater than a & b")
```
