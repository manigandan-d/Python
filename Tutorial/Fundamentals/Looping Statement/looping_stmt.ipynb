looping statement

for

```python
for i in range(5):
    print("python")
```

while

```python
i = 0

while i < 5:
    print("python")
    i += 1
```

for else

```python
data = [1, 3, 5, 7, 9]

# data = [2, 4, 6, 8, 10]

for i in data:
    if i % 2 == 0:
        print(i)
        break

else:
    print("There's no even numbers in the data")
```

while else

```python
data = [1, 3, 5, 7, 9]

# data = [2, 4, 6, 8, 10]

i = 0
n = len(data)

while i < n:
    if data[i] % 2 == 0:
        print(data[i])
        break

    i += 1

else:
    print("There's no even numbers in the data")
```

reversing a range using negative step value

```python
for i in range(5, 0, -1):
    print(i)
```

reversing a range using reversed()

```python
for i in reversed(range(1, 6, 1)):
    print(i)
```
