Check object is iterable or not

```python
def it(obj):
    try:
        iter(obj)
        return True
    
    except TypeError:
        return False
    

data = [5, 2.5, "python", True, [2, 4, 6], (1, 3, 5), {"Name": "Mark", "age": 30}, {"lion", "tiger", "cheetah"}]

for i in data:
    print(f"{i} is iterable: {it(i)}")
```

Iterating on an iterator

```python
object = "computer"

it = iter(object)

for i in object:
    print(it.__next__(), end=" ")
```

```python
evenNum = [2, 4, 6, 8, 10]

it = iter(evenNum)

for i in evenNum:
    print(it.__next__(), end=" ")
```

```python
oddNum = (1, 3, 5, 7, 9)

it = iter(oddNum)

for i in oddNum:
    print(it.__next__(), end=" ")
```

```python
ada_lovelace = {"full_name": "Ada Lovelace", 
                "birth_date": "Dec 10, 1815", 
                "occupation": "Mathematician, Writer", 
                "known_for": "First computer programmer", 
                "achievement": "Pioneering work with Charles Babbage on the Analytical Engine"
                }

it = iter(ada_lovelace)

for i in ada_lovelace:
    print(ada_lovelace[it.__next__()])
```

```python
fruits = {"apple", "banana", "orange", "mango", "grapes"}

it = iter(fruits)

for i in fruits:
    print(it.__next__(), end=" ")
```

Creation of an iterator

```python
class FromTenToLimit:
    def __init__(self, limit):
        self.limit = limit

    def __iter__(self):
        self.x = 10
        return self
    
    def __next__(self):
        val = self.x

        if val > self.limit:
            raise StopIteration

        self.x += 1

        return val
    

values = FromTenToLimit(15)

for i in values:
    print(i, end=" ")

values = FromTenToLimit(5)

for i in values:
    print(i, end=" ")
```

Every iterator is an iterable but not every iterable is an iterator
