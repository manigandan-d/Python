indentation 
* leading whitespace before any statement in python
* indicating a block of code

```python
data = list(range(10))

print(data)

evenData, oddData = [], []

for i in data:
    if i%2 == 0:
        evenData.append(i)
        
    else: 
        oddData.append(i)
        
print(evenData)
print(oddData)
```
