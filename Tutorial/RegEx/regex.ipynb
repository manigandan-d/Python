```python
# findall()

data = "hey md, how is going on, yeah, it is fine and currently i'm learning python"

print(re.findall("is", data, re.I))
```

```python
data = "hello 123 !@$ABC"

print(re.findall("\w", data))
print(re.findall("[a-zA-Z0-9]", data))
print(re.findall("\d+", data))
```

```python
mail = "this is mail id manigandan2261@gmail.com"

res = re.search("\w+\d+@\w+\.\w+", mail)

print(res.group())
```

user: manigandan2261
domain: gmail
extension: com

```python
mail = "this is mail id manigandan2261@gmail.com"

res = re.search("(\w+\d+)@(\w+)\.(\w+)", mail)

print(res.group(1))
print(res.group(2))
print(res.group(3))
```

finditer() - searching in the file 

```python
pattern = re.compile("[^b]at")

with open("demo.txt", "r") as fobj:
    contents = fobj.read()
    
match = pattern.finditer(contents)

print(match)

for i in match:
    print(i, i.group())
```
