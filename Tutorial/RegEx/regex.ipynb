Regular Expression (RegEx) - check if a string contains the specified search pattern or not 

1. search function 
2. match object 
3. metacharacters
4. special sequences 

```python
import re

data = "Hello World!"
```

re.search()

* searches a string for a match
* returns match object if there is a match else None

```python
print(re.search("^Hello", data))
print(re.search("^ello", data))
```

```python
if re.search("^Hello", data):
    print("The given string starts with 'Hello'.")
    
else: 
    print("The given string doesn't start with 'Hello'.")
```
