```python
import re
```

```python
data = "Hello World"

print(re.search("^Hello", data))
print(re.search("^Fello", data))

if re.search("^Hello", data):
    print("The given data starts with Hello")
    
else:
    print("The given data doesn't start with Hello")
    
if re.search("World$", data):
    print("The given data ends with World")
    
else:
    print("The given data doesn't end with World")
```

```python
data = "Hello World"

match = re.search("\s", data)

print(match)
print(match.start())
print(match.end())
```

```python
data = "Python is an object oriented programming language, python is known for its simplicity"

print(re.findall("[Pp]ython", data))
```

```python
data = "hey, this is my phone numer: 1234509876"

print(re.findall("[0-9]", data))
print(re.findall("[0-9]+", data))
print(re.findall("[^a-z]", data))
```

```python
mail = "this is mail id manigandan2261@gmail.com"

res = re.search("\w+\d+@\w+\.\w+", mail)

print(res.group())
```

user: manigandan2261
domain: gmail
extension: com

```python
mail = "this is mail id manigandan2261@gmail.com"

res = re.search("(\w+\d+)@(\w+)\.(\w+)", mail)

print(res.group(1))
print(res.group(2))
print(res.group(3))
```

finditer() - searching in the file 

```python
pattern = re.compile("[^b]at")

with open("demo.txt", "r") as fobj:
    contents = fobj.read()
    
match = pattern.finditer(contents)

print(match)

for i in match:
    print(i, i.group())
```
