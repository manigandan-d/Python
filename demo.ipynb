decorators - built in and user defined 

built in decorators - class, static and property decorators 

class decorator - @classmethod, explicit call, access , bounded to obj
  #access only class attribute
  #similar to normal method -explicit call
  #implement  - @classmethod
  #bounded to object

```python
class Demo:
    cnt = 100 # class attribute 
    
    def get(self): # not-static method 
        self.var = 350
        print(self.var)
        print(Demo.cnt)
    
    @classmethod
    def classFun(self, msg): # class decorator 
        print(Demo.cnt)
        print(msg)
        # print(self.var) - can't access instance attribute 
    
d = Demo()
d.get()
d.classFun("md")
```

Static decorator
  #access only class attribute
  #similar to normal method -explicit call
  #implement  - @staticmethod
  #not bounded to object
  #ex:logging ,Environment,....etc

```python
# static decorator - @staticmethod - explicit call, access , not bounded to obj

class Demo:
    cnt = 100 # class attribute 
    
    def get(self): # not-static method 
        self.var = 350
        print(self.var)
        print(Demo.cnt)
    
    @staticmethod
    def classFun(msg): # class decorator 
        print(Demo.cnt)
        print(msg)
        # print(self.var) - can't access instance attribute 
    
d = Demo()
d.get()
Demo.classFun("md") # we can use _> d.classFun("md") but it's meaningless 
```

property decorator

1. getter
2. setter
3. deleter

```python
class Employee:
  def __init__(self, fname, lname, id):
    self.fname = fname
    self.lname = lname
    self.id = id

  @property
  def email(self):
    return self.fname+self.lname+"@gmail.com"

  @property
  def fullname(self): # getter
    return self.fname+" "+self.lname

  @fullname.setter
  def fullname(self, name):
    self.fname, self.lname = name.split()

  @fullname.deleter
  def fullname(self):
    self.fname = None
    self.lname = None
    self.id = None

    print("fullname is deleted")

fname = "anbu"
lname = "mani"
id = 12345

emp1 = Employee(fname, lname, id)
print(emp1.email)
print(emp1.fullname)

# how can we change anbu into bala

# fname = "bala" -> first name
# emp2 = Employee(fname, lname, id)
# print(emp2.email())
# print(emp2.fullname())

# emp1.fname = "bala" -> second way

emp1.fullname = "bala mani"
print(emp1.email)
print(emp1.fullname)

del emp1.fullname
```

closure & decorator 
* declare an outer function - this is called as decorator 
* decorator take only one argument
* declare a inner function inside the decorator 
* inner should return result 
* outer function should return the inner function 

closure 
* technique where outer function holds the value even after going out of scope 

```python
def pow(temp):
    def inner(a, b):
        return temp(a, b)**2
    
    return inner

def add(a, b):
    return a+b

def mul(a, b):
    return a*b

a=5
b=2

clofun1 = pow(add)
print(clofun1(a, b))

clofun2 = pow(mul)
print(clofun2(a, b))
```

decorator

```python
def pow(temp):
    def inner(a, b):
        return temp(a, b)**2
    
    return inner

@pow
def add(a, b):
    return a+b

@pow
def mul(a, b):
    return a*b

a = 5
b = 2

print(add(a, b))
print(mul(a, b))
```

```python
def minion_game(data):
    data = data.lower()
    data_len = len(data)
    stuart_words = {}
    kevin_words = {}
    vowels = ['a', 'e', 'i', 'o', 'u']

    for i in range(data_len):
        for j in range(i+1, data_len+1):
            substring = data[i:j]
            if substring[0] in vowels:
                kevin_words[substring] = kevin_words.get(substring, 0) + 1
            else:
                stuart_words[substring] = stuart_words.get(substring, 0) + 1

    stuart_score = sum(stuart_words.values())
    kevin_score = sum(kevin_words.values())

    if stuart_score > kevin_score:
        print(f"Stuart {stuart_score}")
    elif kevin_score > stuart_score:
        print(f"Kevin {kevin_score}")
    else:
        print("Draw")

if __name__ == '__main__':
    s = input()
    minion_game(s)
```

```python
def minion_game(data):
    data_len = len(data)
    stuart_score = 0
    kevin_score = 0
    vowels = set(['A', 'E', 'I', 'O', 'U'])

    for i in range(data_len):
        if data[i] in vowels:
            # All substrings starting from this vowel are unique
            kevin_score += data_len - i
        else:
            # All substrings starting from this consonant are unique
            stuart_score += data_len - i

    if stuart_score > kevin_score:
        print(f"Stuart {stuart_score}")
    elif kevin_score > stuart_score:
        print(f"Kevin {kevin_score}")
    else:
        print("Draw")

if __name__ == '__main__':
    s = input().strip()
    minion_game(s)
```
