property decorator

1. getter
2. setter
3. deleter

```python
class Employee:
  def __init__(self, fname, lname, id):
    self.fname = fname
    self.lname = lname
    self.id = id

  @property
  def email(self):
    return self.fname+self.lname+"@gmail.com"

  @property
  def fullname(self): # getter
    return self.fname+" "+self.lname

  @fullname.setter
  def fullname(self, name):
    self.fname, self.lname = name.split()

  @fullname.deleter
  def fullname(self):
    self.fname = None
    self.lname = None
    self.id = None

    print("fullname is deleted")

fname = "anbu"
lname = "mani"
id = 12345

emp1 = Employee(fname, lname, id)
print(emp1.email)
print(emp1.fullname)

# how can we change anbu into bala

# fname = "bala" -> first name
# emp2 = Employee(fname, lname, id)
# print(emp2.email())
# print(emp2.fullname())

# emp1.fname = "bala" -> second way

emp1.fullname = "bala mani"
print(emp1.email)
print(emp1.fullname)

del emp1.fullname
```
