decorators - built in and user defined 

built in decorators - class, static and property decorators 

class decorator - @classmethod, explicit call, access , bounded to obj

```python
class Demo:
    cnt = 100 # class attribute 
    
    def get(self): # not-static method 
        self.var = 350
        print(self.var)
        print(Demo.cnt)
    
    @classmethod
    def classFun(self, msg): # class decorator 
        print(Demo.cnt)
        print(msg)
        # print(self.var) - can't access instance attribute 
    
d = Demo()
d.get()
d.classFun("md")
```

Static decorator

```python
# static decorator - @staticmethod - explicit call, access , not bounded to obj

class Demo:
    cnt = 100 # class attribute 
    
    def get(self): # not-static method 
        self.var = 350
        print(self.var)
        print(Demo.cnt)
    
    @staticmethod
    def classFun(msg): # class decorator 
        print(Demo.cnt)
        print(msg)
        # print(self.var) - can't access instance attribute 
    
d = Demo()
d.get()
Demo.classFun("md") # we can use _> d.classFun("md") but it's meaningless 
```

property decorator

1. getter
2. setter
3. deleter

```python
class Employee:
  def __init__(self, fname, lname, id):
    self.fname = fname
    self.lname = lname
    self.id = id

  @property
  def email(self):
    return self.fname+self.lname+"@gmail.com"

  @property
  def fullname(self): # getter
    return self.fname+" "+self.lname

  @fullname.setter
  def fullname(self, name):
    self.fname, self.lname = name.split()

  @fullname.deleter
  def fullname(self):
    self.fname = None
    self.lname = None
    self.id = None

    print("fullname is deleted")

fname = "anbu"
lname = "mani"
id = 12345

emp1 = Employee(fname, lname, id)
print(emp1.email)
print(emp1.fullname)

# how can we change anbu into bala

# fname = "bala" -> first name
# emp2 = Employee(fname, lname, id)
# print(emp2.email())
# print(emp2.fullname())

# emp1.fname = "bala" -> second way

emp1.fullname = "bala mani"
print(emp1.email)
print(emp1.fullname)

del emp1.fullname
```

closure & decorator 
* declare an outer function - this is called as decorator 
* decorator take only one argument
* declare a inner function inside the decorator 
* inner should return result 
* outer function should return the inner function 

closure 
* technique where outer function holds the value even after going out of scope 

```python
def pow(temp):
    def inner(a, b):
        return temp(a, b)**2
    
    return inner

def add(a, b):
    return a+b

def mul(a, b):
    return a*b

a=5
b=2

clofun1 = pow(add)
print(clofun1(a, b))

clofun2 = pow(mul)
print(clofun2(a, b))
```

decorator

```python
def pow(temp):
    def inner(a, b):
        return temp(a, b)**2
    
    return inner

@pow
def add(a, b):
    return a+b

@pow
def mul(a, b):
    return a*b

a = 5
b = 2

print(add(a, b))
print(mul(a, b))
```

Single Inheritance

```python
class Employee:
    def __init__(self, name, id):
        self.__name = name
        self.__id = id
        
        print("this is am employee")
        
    def name(self):
        return self.__name
    
    def id(self):
        return self.__id

class Developer(Employee):
    def __init__(self, skill):
        super().__init__("mani", 12345)
        self.__skill = skill
        
        print("this is a developer")
        
    def display(self):
        print(super().name())
        print(super().id())
        print(self.__skill)

dev = Developer("python")
dev.display()
```

Operator Overloading
* giving special meaning to the operator 

```python
class Circle: 
    def __init__(self, rad):
        self.__rad = rad
        
    def __str__(self):
        return f"{self.__rad}"
    
    def __add__(self, other):
        return self.__rad + other.__rad 
    
c1 = Circle(5)
c2 = Circle(3)

print(c1) # __str__()

c3 = c1+c2

print(c3)

d1 = Circle("anbu")
d2 = Circle("mani")

d3 = d1+d2

print(d3)
```

Complex number

```python
class ComplexNum:
    def __init__(self, real, img):
        self.__real = real
        self.__img = img
        
    def __str__(self):
        return f"{self.__real} + {self.__img}i"
    
    def __add__(self, other):
        real = self.__real + other.__real
        img = self.__img + other.__img
        
        # return f"{real} + {img}i"
        return ComplexNum(real, img)
        
comp1 = ComplexNum(3, 2)
comp2 = ComplexNum(5, 1)

print(comp1)
print(comp2)

comp3 = comp1 + comp2 
print(comp3)
```

Comparison 

```python
class Employee:
    def __init__(self, name, exp):
        self.__name = name
        self.__exp = exp 
        
    def __gt__(self, other):
        if self.__exp > other.__exp:
            return 1 
        
        else: 
            return 0
        
emp1 = Employee("mani", 25)
emp2 = Employee("anbu", 10)

if emp1 > emp2:
    print("Mani has higher experience")
    
else: 
    print("Anbu has higher experience")
```
