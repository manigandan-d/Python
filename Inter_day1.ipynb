```python
#Class Attribute /static Attribute
```

```python
# They are declared inside the class
# They are "Shared"  and "Accessed" all the elements of class
# These attributes are Accessed with "ClassName".
# memory of the class attribute are shared among the instance of class
```

```python
class Employee:
  empName="Sam"    #class Attribute
  empId=987654     #class scope

  def getInfo(self): #Non-static method
    print("EmpName-get:",Employee.empName)
    print("EmpId-get:",Employee.empId)

  def disInfo(self):
    print("EmpName-dis:",Employee.empName)
    print("EmpId-disp:",Employee.empId)


emp1=Employee()
emp1.getInfo()
emp1.disInfo()
print("EmpName:",Employee.empName)
print("EmpID:",Employee.empId)
```

```python
class Employee:
  empName="Sam"    #class Attribute
  empId=987654     #class scope

  def getInfo(self): #Non-static method
    empName="Subha"
    print("EmpName-get:",empName)   #Local
    print("EmpId-get:",Employee.empId)

  def disInfo(self):
    print("EmpName-dis:",Employee.empName)
    print("EmpId-disp:",Employee.empId)


emp1=Employee()
emp1.getInfo()
emp1.disInfo()
print("EmpName:",Employee.empName)
print("EmpID:",Employee.empId)
```

```python
class Employee:
  empid=100                    #Class Attribute -heap
  def eidGen(self):            #class attribute memory -shared among all the instance
    Employee.empid+=1

  def dispId(self):
    print("Eid:",Employee.empid)


emp1=Employee()              #emp1  =>empid  =>@2000 =>100
emp1.eidGen()                       #=>@2000  =>101
#emp1.dispId()                 #101

emp2=Employee()               #emp2  =>empid =>@2000  =>101
emp2.eidGen()                         #=>@2000  =>102
#emp2.dispId()                  #101 102  --->last executed value

emp2.dispId()
emp1.dispId()
```

```python
# Instance Attribute
```

```python
# instance Scope
#  memory of the Instance attribute are not shared among the instance of class
# every Object will have a own copy of attribute
#Accessed with instance name
```

```python
class Employee:

  def getInfo(self): #Non-static method
    self.empName=input("Enter Name:")   #instance
    self.empId=input("Enter Id:")
    print("EmpName-get:",self.empName)
    print("EmpId-get:",self.empId)

  def disInfo(self):
    print("EmpName-dis:",self.empName)
    print("EmpId-disp:",self.empId)


emp1=Employee()   #empName="SAM"     empID=
emp1.getInfo()
emp1.disInfo()
print()
emp2=Employee()   #empName="RAM"    empID=
emp2.getInfo()
emp2.disInfo()
```

```python
class Employee:
  #empid=100                    #Class Attribute -heap
  def eidGen(self):            #class attribute memory -shared among all the instance
    #Employee.empid+=1
    self.empid=100
    self.empid+=1

  def dispId(self):
    print("Eid:",self.empid)


emp1=Employee()              #emp1  =>empid  =>@2000 =>100
emp1.eidGen()                       #=>@2000  =>101
                #101

emp2=Employee()               #emp2  =>empid =>@4000  =>100
emp2.eidGen()                         #=>@4000  =>101

emp2.dispId()                  #101
emp1.dispId()
```
