counts the number of digits, letters and special characters from the given string

```python
data = input("Enter the data: ")

num_digits, num_letters, num_special = 0, 0, 0

for char in data:
    if char.isdigit():
        num_digits += 1
        
    elif char.isalpha():
        num_letters += 1
        
    else:
        num_special += 1
        
print(f"Number of digits in {data} is {num_digits}")
print(f"Number of letters in {data} is {num_letters}")
print(f"Number of special characters in {data} is {num_special}")
```
