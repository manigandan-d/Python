The Minion Game is a word game played between two players, traditionally named Stuart and Kevin. The game is inspired by the movie "Despicable Me" and its characters, the Minions. Here are the rules of the Minion Game:

Setup:
* Both players are given the same string, which can be a word, phrase, or sequence of letters.
* The string consists of uppercase letters only.

Objective:
* The goal of the game is to score points by creating substrings using the letters of the given string.

Game Rules:
* Stuart and Kevin take turns making substrings using the letters of the string.
* Stuart starts the game and makes words starting with consonants (non-vowels).
* Kevin follows Stuart's turn and makes words starting with vowels (A, E, I, O, U).
* Each player can create substrings using consecutive letters from the original string.
* The game continues until all possible substrings have been created.

Scoring:
* Each player earns points based on the number of words they create and the words they create.
* A player gets +1 point for each occurrence of the substring in the original string.
* Stuart scores points for words that start with consonants.
* Kevin scores points for words that start with vowels.

Determining the Winner:
* At the end of the game, the scores of Stuart and Kevin are compared.
* The player with the higher score wins the game.
* If both players have the same score, the game is a draw.

```python
def minion_game(data):
    data = data.strip()
    
    vowels = "AEIOU"
    
    stuart_score = 0
    kevin_score = 0
    
    length = len(data)

    for i in range(length):
        if data[i] in vowels:
            kevin_score += length - i
            
        else:
            stuart_score += length - i

    if stuart_score > kevin_score:
        print("Stuart", stuart_score)
        
    elif kevin_score > stuart_score:
        print("Kevin", kevin_score)
        
    else:
        print("Draw")

if __name__ == '__main__':
    s = input()
    minion_game(s)
```
