```python
cnt=100                         #Global Attribute
class Employee:
  def getInfo(self):
    print("Gobal-cnt get:",cnt)

  def putInfo(self):
    print("Gobal-cnt put:",cnt)
    print("Gobal():",globals()['cnt'])

emp=Employee()
emp.getInfo()
emp.putInfo()
print("Global Attribute:",cnt)
```

```python
cnt=100                         #Global Attribute
class Employee:
  def getInfo(self):
    cnt=50                       #Local Attribute
    print("Gobal-cnt get:",cnt)      # 50

  def putInfo(self):
    print("Gobal-cnt put:",cnt)        # 100
    print("Gobal():",globals()['cnt']) # 100

emp=Employee()
emp.getInfo()
emp.putInfo()
print("Global Attribute:",cnt)        #100
```

```python
cnt=100                         #Global Attribute
class Employee:
  def getInfo(self):
    global cnt
    cnt=50
    print("Gobal-cnt get:",cnt)      # 50

  def putInfo(self):
    print("Gobal-cnt put:",cnt)        # 50
    print("Gobal():",globals()['cnt']) # 50

emp=Employee()
emp.getInfo()
#emp.putInfo()
print("Global Attribute:",cnt)        #50
```

```python
#Class Decorator
  #access only class attribute
  #similar to normal method -explicit call
  #implement  - @classmethod
  #bounded to object

 #static Decorator
  #access only class attribute
  #similar to normal method -explicit call
  #implement  - @staticmethod
  #not bounded to object
  #ex:logging ,Environment,....etc
```

```python
class Sample:
  cnt=250          # Class Attribute
  def get(self):   #non-static method
    self.var=350
    print("Var:",self.var)
    print("Cnt:",Sample.cnt)

  @classmethod
  def classfun(self,msg):     #class
    print("Count:",Sample.cnt) #class
    print("Msg:",msg)          #Local
    #print("Vaiabl:",self.var)    error

  @staticmethod
  def classStatic(msg):
    print("Static Count:",Sample.cnt)
    print("Static Msg:",msg)
    #print("Vaiabl:",self.var)  error

obj=Sample()
obj.get()
obj.classfun("C The Change!!!")  #classfun(obj,".....")
Sample.classStatic("Hi all")      #className.method
obj.classStatic("Hi all")        #???? why??
```

```python
class MyEmployee:
  def __init__(self,fname,lname,sal):   #fname,lname,sal= local
    self.fname=fname
    self.lname=lname
    self.sal=sal

  @property
  def email(self):
    return self.fname+'.'+self.lname+'@mail.com'

  @property
  def fullname(self):  #getter
    return self.fname+' '+self.lname

  @fullname.setter
  def fullname(self,Name):  #Name="Arun Kumar"
    self.fname,self.lname=Name.split(' ')

  @fullname.deleter
  def fullname(self):
    self.fname=None
    self.lname=None
    self.sal=None
    print("FullName Deleted")

```
